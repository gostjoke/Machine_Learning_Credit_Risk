```python
# importing libraries
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # essential graphical output
import seaborn as sns # enhanced graphical output

df = pd.read_csv("credit_risk_training.csv")
df.head(15)
```

```python
#use map to fix
d = {'no_checking': '0', 'less_0': '10' ,'0_to_200': '100', 'greater_200': '200'}
df["CheckingStatus"] = df["CheckingStatus"].map(d)
df[['CheckingStatus']] = df[['CheckingStatus']].apply(pd.to_numeric)
# fix telephone and risk
d1 = {'none':'0', 'yes':'1'}
d2 = {'No Risk':'0', 'Risk':'1'}
df["Telephone"] = df["Telephone"].map(d1)
df["Risk"] = df["Risk"].map(d2)

df[['Telephone', 'Risk']] = df[['Telephone','Risk']].apply(pd.to_numeric)

# EmploymentDuration
d3 = {'unemployed':'0', 'less_1':'1', '1_to_4':'2.5', '4_to_7': '5.5', 'greater_7':'10'}
df["EmploymentDuration"] = df["EmploymentDuration"].map(d3)
#sex
d4 = {'female':'0', 'male':"1"}
df["Sex"] = df["Sex"].map(d4)

df[['EmploymentDuration', "Sex"]] = df[['EmploymentDuration','Sex']].apply(pd.to_numeric)

### housing 
d5 = {"free":'1', 'own':'1', 'rent':'0'}
df["Housing"] = df["Housing"].map(d5)
###OthersOnLoan
d6 = {"none":"0", "co-applicant":"1", "guarantor":"2"}
df["OthersOnLoan"] = df["OthersOnLoan"].map(d6)

df[['Housing', "OthersOnLoan"]] = df[['Housing','OthersOnLoan']].apply(pd.to_numeric)

## ExistingSavings
d7 = {'unknown':'0' , 'less_100':'100', '100_to_500':'300', '500_to_1000':'750', 'greater_1000':'1000'}
df["ExistingSavings"] = df["ExistingSavings"].map(d7)
df[["ExistingSavings"]] = df[['ExistingSavings']].apply(pd.to_numeric)


df.head(20)


```

```python
df.to_excel("risk_fixed.xlsx")
```

```python
df.corr(method='pearson')
```

```python
X = df[['CheckingStatus', 'LoanDuration', 'LoanAmount', 'ExistingSavings', 'EmploymentDuration', 'InstallmentPercent',
        'Sex', 'OthersOnLoan', 'Age', 'Housing','ExistingCreditsCount', 'Dependents','Telephone']]
y = df['Risk']
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
#'CheckingStatus' :100, 'LoanDuration' 20, 'LoanAmount' 3693, 'EmploymentDuration' 5.5, 'InstallmentPercent' 3,
#'Sex' 0, 'OthersOnLoan' 1, 'Age' 41, 'Housing' 0,'ExistingCreditsCount' 2, 'Dependents' 1,'Telephone' 1
predictedRisk = regr.predict([[100, 20, 3000, 600 ,5.5, 3, 0, 1, 41, 0, 2, 1, 1]])
print(predictedRisk)
```
